```python
c = CurrencyConverter()


def sale(text):
    result = 0  # Задаем значение по умолчанию
    if text.split(" ")[1] == "руб.,":
        result = int(text.split(" ")[0])
    elif text.split(" ")[1] == "$,":
        result = int(c.convert((text.split(" ")[0]), "USD", "RUB"))
    elif text.split(" ")[1] == "€,":
        result = int(c.convert((text.split(" ")[0]), "EUR", "RUB"))
    return result
```

```python
text = df["Цена"]
text = text.apply(sale)
df["Цена"] = text
```

```python
from geopy.exc import GeocoderTimedOut
from geopy.distance import geodesic
from geopy.geocoders import Nominatim
```

```python
empty_list = []


def do_geocode(address, attempt=1, max_attempts=5):
    try:
        empty_list.append(
            Nominatim(user_agent="my_request").geocode(address, timeout=None)[-1]
        )
    except GeocoderTimedOut:
        if attempt <= max_attempts:
            return do_geocode(address, attempt=attempt + 1)
        raise
```
