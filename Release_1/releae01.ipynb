```python
df = df.drop_duplicates(subset=["ID  объявления"], keep='first').reset_index(drop=True)
```

```python
df.info()
```

```python
df = df.drop(
    ["Unnamed: 0", "Описание", "Название ЖК", "Серия дома", "Ссылка на объявление"], axis=1
)
```

```python
df.info()
```

```python
df["Количество комнат"] = df["Количество комнат"].fillna('Свободная планировка')
df['Ремонт'] = df['Ремонт'].fillna('Без ремонта')
```

```python
df['Количество комнат'].unique()
```

```python
df["Тип квартиры"] = [
    row.split(", ")[-1] if len(row) != 1 else "Стандарт"
    for row in df["Количество комнат"]
]
```

```python
df.head()
```

```python
df["Количество комнат"] = [
    int(row.split(", ")[0]) if len(row.split(", ")[0]) == 1 else 0
    for row in df["Количество комнат"]
]
```

```python
df["Метро"].unique()
```

```python
df['Метро'] = df['Метро'].astype(str)
```

```python
df["Метро"].dtype
```

```python
df["Метро"] = [row.split(" (")[0] for row in df["Метро"]]
df["Метро"] = [row.split(". ")[1] if ". " in row else row for row in df["Метро"]]
```

```python
df['Метро']
```

```python
df["Площадь кв, м2"] = [row.split("/")[0] for row in df["Площадь, м2"]]
```

```python
df = df.drop(["Площадь, м2"], axis=1)
df = df.drop(["Площадь комнат, м2"], axis=1)
```
