```python
df['Балкон'].unique()
```

```python
df["Балкон"] = df["Балкон"].fillna("Балкон (0), Лоджия (0)")
df["Лоджия"] = [
    row.split(" ")[-1] if (row.split(" ")[-2]) == "Лоджия" else "(0)"
    for row in df["Балкон"]
]
df["Балкон"] = [
    row.split(" ")[1] if row.split(" ")[0] == "Балкон" else "(0)"
    for row in df["Балкон"]
]
```

```python
df["Лоджия"] = [row.split("(")[1] for row in df["Лоджия"]]
df["Лоджия"] = [int(row.split(")")[0]) for row in df["Лоджия"]]
df["Балкон"] = [row.split("(")[1] for row in df["Балкон"]]
df["Балкон"] = [int(row.split(")")[0]) for row in df["Балкон"]]
```

```python
df["Лифт"] = [
    (
        row
        if row == row
        else [
            (
                "Пасс (1), Груз (1)"
                if df["Этажность дома"].iloc[i] > 5
                else "Пасс (0), Груз (0)"
            )
        ][0]
    )
    for i, row in enumerate(df["Лифт"])
]
```

```python
df["Лифт пасс"] = [
    row.split(" ")[1] if (row.split(" ")[0]) == "Пасс" else "(0)" for row in df["Лифт"]
]
df["Лифт груз"] = [
    row.split(" ")[-1] if (row.split(" ")[-2]) == "Груз" else "(0)"
    for row in df["Лифт"]
]
```

```python
df["Лифт пасс"] = [row.split("(")[1] for row in df["Лифт пасс"]]
df["Лифт пасс"] = [int(row.split(")")[0]) for row in df["Лифт пасс"]]
df["Лифт груз"] = [row.split("(")[1] for row in df["Лифт груз"]]
df["Лифт груз"] = [int(row.split(")")[0]) for row in df["Лифт груз"]]
df = df.drop(["Лифт"], axis=1)
```

```python
df["Санузел"] = df["Санузел"].fillna("Совмещенный (1)")
df["Санузел совмещенный"] = [
    row.split(" ")[1] if (row.split(" ")[0]) == "Совмещенный" else "(0)"
    for row in df["Санузел"]
]
df["Санузел раздельный"] = [
    row.split(" ")[-1] if (row.split(" ")[-2]) == "Раздельный" else "(0)"
    for row in df["Санузел"]
]
df["Санузел совмещенный"] = [row.split("(")[1] for row in df["Санузел совмещенный"]]
df["Санузел совмещенный"] = [
    int(row.split(")")[0]) for row in df["Санузел совмещенный"]
]
df["Санузел раздельный"] = [row.split("(")[1] for row in df["Санузел раздельный"]]
df["Санузел раздельный"] = [int(row.split(")")[0]) for row in df["Санузел раздельный"]]
df = df.drop(["Санузел"], axis=1)
```

```python
c = CurrencyConverter()


def sale(text):
    result = 0  # Задаем значение по умолчанию
    if text.split(" ")[1] == "руб.,":
        result = int(text.split(" ")[0])
    elif text.split(" ")[1] == "$,":
        result = int(c.convert((text.split(" ")[0]), "USD", "RUB"))
    elif text.split(" ")[1] == "€,":
        result = int(c.convert((text.split(" ")[0]), "EUR", "RUB"))
    return result
```

```python
text = df["Цена"]
text = text.apply(sale)
df["Цена"] = text
```

```python
def plot_mean_price(column):
    plt.figure(figsize=(8, 4)) # определение размера графиков
    n_bars = np.arange(len(df_msk[column].unique())) # массив с кол-во уникальных значений в каждой катег колонке.
    
    # создание дф по каждой колонке групируя категории по средней цене
    grouped = pd.DataFrame(df_msk.groupby([column])['Price'].mean().round())
    
    plt.title(column) #  заголовок
    plt.bar(n_bars, grouped['Price']) # создание графика
    plt.xticks(n_bars, grouped.index); # подписываем колонки
    plt.show() # чтобы не накладывался на один график
```

```python
Q1 = df_msk["Ceiling height, m"].quantile(0.25)
Q3 = df_msk["Ceiling height, m"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


filtered_df = df_msk[
    (df_msk["Ceiling height, m"] >= lower_bound)
    & (df_msk["Ceiling height, m"] <= upper_bound)
]

plt.figure(figsize=(10, 6))
plt.scatter(
    filtered_df["ID"],
    filtered_df["Ceiling height, m"],
    color="skyblue",
    edgecolor="black",
)
plt.title("Распределение высоты потолков в квартирах")
plt.xlabel("ID квартиры")
plt.ylabel("Высота потолков, м")
plt.grid(True)
plt.show()
```
